```python
import sqlite3
```

```python
dbpath = 'data/ecmdatabase.db'
con = sqlite3.connect(f"file:{dbpath}?mode=ro", uri=True)
with con:
    result = con.execute("SELECT * from companies WHERE stock_symbol = 'AAPL';")
    records = result.fetchall()
    company_name = records[0][1]
    apple_item1 = records[0][2]
```

```python
apple_item1 = apple_item1.replace('\n', '')
apple_item1
```

```python
from refined.data_types.base_types import Span
from refined.inference.processor import Refined


refined = Refined.from_pretrained(model_name='wikipedia_model',
                                  entity_set="wikipedia")

text = apple_item1
spans = refined.process_text(text)

print(spans)
```

```python
def find_apple_entities(spans):
    apple_related = []
    for span in spans:
        entity_name = span.text  
        entity_info = span.predicted_entity 
        if "Apple" in entity_name:
            apple_related.append(span)
        elif entity_info is not None and hasattr(entity_info, 'wikipedia_entity_title'):
            if entity_info.wikipedia_entity_title and "Apple" in entity_info.wikipedia_entity_title:
                apple_related.append(span)
    
    return apple_related

apple_entities = find_apple_entities(spans)
for entity in apple_entities:
    print(entity.text, entity.predicted_entity)
```
